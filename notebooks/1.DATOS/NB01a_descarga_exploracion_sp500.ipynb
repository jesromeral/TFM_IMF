En este notebook se realiza la descarga, exploración inicial y preparación básica de los datos OHLCV del índice S&P 500 para su uso en los modelos del TFM



```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

plt.style.use("seaborn-v0_8") 
```

```python
os.makedirs("../../datos", exist_ok=True)

DATA_OUT_PATH = "../../datos/sp500_ohlcv_2010_2025.csv"
```

```python
# Descarga de datos del S&P 500 (2010-2025) 
sp500 = yf.download("^GSPC", start="2010-01-01", end="2025-11-30")
sp500.head()
```

```python
sp500.info()

sp500.describe()
```

```python
sp500.isna().sum()
```

```python
plt.figure(figsize=(12,5))
plt.plot(sp500["Close"])
plt.title("Índice S&P 500 - Precio de cierre diario")
plt.xlabel("Fecha")
plt.ylabel("Precio de cierre")
plt.show()
```

```python
sp500.to_csv(DATA_OUT_PATH)

if os.path.exists(DATA_OUT_PATH): 
    print(f"Archivo guardado correctamente en {DATA_OUT_PATH}") 
else: 
    print("Error: el archivo no se ha guardado.")
```
